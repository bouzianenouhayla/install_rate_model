```python
transformations_metadata = {
    "numerical": {
        "columns": numerical_columns,
        "transformations": []
    },
    "list_features": {
        "columns": list_features,
        "transformations": ["aggergate_impressions"]  ##
    },
    "dict_features": {
        "columns": dict_features,
        "transformations": ["expand_all_column"]  ## 
    },

     "aggregated_features": {
        "columns": aggregated_features,
        "transformations": ["expand_top_apps"]  ##
    },

    "cat_features": {
        "columns": cat_features,
        "transformations": ["encoder"]  ##""
    }
    
}

function_map = {
   
    "expand_all_column": ExpandDictStep(),
    "expand_top_apps": JSONToFeaturesStep(top_n=50),
    "aggergate_impressions": AggregateImpressionsStep(),
    "encoder": CategoricalEncodingStep()
   
}

pipeline = TransformationPipeline()

for group_name, group_info in transformations_metadata.items():
    for transformation_name in group_info["transformations"]:
        transformation = function_map[transformation_name]
        if hasattr(transformation, "fit_transform"):  # Handles Scikit-learn or similar transformers
            pipeline.add_step(
                f"{transformation_name}_{group_name}",
                transformation,
                group_info["columns"]
            )
        else:  # Handles custom function-based transformations
            for column in group_info["columns"]:
                pipeline.add_step(
                    f"{transformation_name}_{column}",
                    FunctionStep(transformation),
                    [column]
                )
```

```python
y_prob = model.predict_proba(train_x)[:, 1]  # Get probability of class 1 (install)

# Calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(train_y, y_prob)

# Plot precision-recall curve
plt.plot(recall, precision, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall curve')
plt.show()
```

```python
f1_scores = 2 * (precision * recall) / (precision + recall)
best_threshold = thresholds[np.argmax(f1_scores)]
print("Optimal Threshold for Best F1-Score:", best_threshold)
```

```python
# Predict labels based on a custom threshold
y_pred_custom = (y_prob >= best_threshold).astype(int)

# Evaluate the classification report
print(classification_report(train_y, y_pred_custom))

# Confusion matrix
print(confusion_matrix(train_y, y_pred_custom))
```
