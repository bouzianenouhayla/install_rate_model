## Timestamp data

to understand the differences between the timestamp lets plot the differences between them

```python
timestamp_columns
```

```python
# Convert timestamps to datetime
df['bid_timestamp'] = pd.to_datetime(df['bid_timestamp'])
df['session_start_date'] = pd.to_datetime(df['session_start_date'])
df['install_date'] = pd.to_datetime(df['install_date'])
df['previous_session_start_date'] = pd.to_datetime(df['previous_session_start_date'])
```

```python
def plot_difference(df, col1,col2):
    # Calculate differences in seconds
    df = df.copy()
    df['time_diff_seconds'] = (df[col1] - df[col2]).dt.total_seconds()
    # Calculate IQR (Interquartile Range)
    Q1 = df['time_diff_seconds'].quantile(0.25)
    Q3 = df['time_diff_seconds'].quantile(0.75)
    IQR = Q3 - Q1
    # Define outlier thresholds (1.5 times IQR)
    lower_bound = Q1 - 10* IQR
    upper_bound = Q3 + 10 * IQR

    # Remove outliers
    df_cleaned = df[(df['time_diff_seconds'] >= lower_bound) & (df['time_diff_seconds'] <= upper_bound)]
    # Plot the differences
    plt.figure(figsize=(8, 5))
    plt.hist(df_cleaned['time_diff_seconds'], bins=50, color='skyblue', edgecolor='black')
    plt.title("Distribution of Time Differences in Seconds")
    plt.xlabel("Time Difference (seconds)")
    plt.ylabel("Frequency")
    plt.xlim(df_cleaned['time_diff_seconds'].min(), df_cleaned['time_diff_seconds'].max()) 
    plt.show()
```

```python
plot_difference(df, 'bid_timestamp','session_start_date')
```

```python
plot_difference(df,'session_start_date', 'previous_session_start_date')
```

```python
plot_difference(df,'session_start_date', 'install_date')
```

```python
plot_difference(df,'bid_timestamp', 'install_date')
```
